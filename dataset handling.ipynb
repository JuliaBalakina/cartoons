```python
from nltk.tokenize import word_tokenize

nltk.download('punkt') # For tokenizer
nltk.download('punkt_tab')
# This downloads the pre-trained Russian tagger data
nltk.download('tagsets') 
nltk.download('averaged_perceptron_tagger_rus') 

def extract_russian_noun_phrases(text):
    # Tokenize and POS tag the Russian text using NLTK's built-in tagger
    tokens = word_tokenize(text, language='russian')
    tagged_tokens = nltk.pos_tag(tokens, lang='rus')

    # Define a simple grammar for noun phrases (e.g., Adjective(s) + Noun(s))
    # 'A' is the RNC tag for Adjective, 'S' for Noun
    grammar = r"""
      NP: {<A>*<S>+}   # Chunk sequences of Adjectives followed by Nouns
    """
    cp = nltk.RegexpParser(grammar)
    tree = cp.parse(tagged_tokens)

    noun_phrases = []
    for subtree in tree.subtrees():
        if subtree.label() == 'NP':
            # Reconstruct the noun phrase from the words in the subtree
            phrase = " ".join([word for word, tag in subtree.leaves()])
            noun_phrases.append(phrase)
    return noun_phrases

# Example text: "у меня было красивое и нежное лицо" (I had a beautiful and delicate face)

df_ru["chunks"] = df_ru.text.apply(extract_russian_noun_phrases)
# Expected output might be: ['красивое и нежное лицо']
df_ru.chunks[0]
```
